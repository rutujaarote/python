```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 8, 5]

# Line Plot
def line_plot():
    plt.plot(x, y, label="sin(x)")
    plt.plot(x, z, label="cos(x)")
    plt.title("Line Plot")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.legend()
    plt.show()

# Bar Plot
def bar_plot():
    plt.bar(categories, values)
    plt.title("Bar Plot")
    plt.xlabel("Categories")
    plt.ylabel("Values")
    plt.show()

# Scatter Plot
def scatter_plot():
    plt.scatter(x, y, label="sin(x)", color="blue")
    plt.scatter(x, z, label="cos(x)", color="red")
    plt.title("Scatter Plot")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.legend()
    plt.show()

# Histogram
def histogram():
    data = np.random.randn(1000)
    plt.hist(data, bins=30, color="purple", alpha=0.7)
    plt.title("Histogram")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()

# Pie Chart
def pie_chart():
    plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'pink'])
    plt.title("Pie Chart")
    plt.show()

# Display all plots
if __name__ == "__main__":
    print("Line Plot:")
    line_plot()
    
    print("Bar Plot:")
    bar_plot()
    
    print("Scatter Plot:")
    scatter_plot()
    
    print("Histogram:")
    histogram()
    
    print("Pie Chart:")
    pie_chart()
```
